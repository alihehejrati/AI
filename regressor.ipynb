```python
from sympy import symbols
import numpy as np
from sympy.plotting import plot as symplot
```

```python
class Polynomial:
    __Px = None # class attribute

    def __init__(self, points, f):
      self.__Px = self.polynomial(points, f)

    def polynomial(self, points, f):
      d = points.shape[0]
      N = points.shape[1]
      x = symbols('x')
      Px = 0

      for i in range(N):
        Pi_x = 1
        for j in range(N):
          if i == j:
            continue
          Pi_x = Pi_x * ((x-points[:, j]) / (points[:, i]-points[:, j]))
        Pi_x = Pi_x * f[i]
        Px = Px + Pi_x
      
      return Px

    def eval(self, x):
      print(self.__Px)
      return sym.sympify(self.__Px).subs({'x': x})[0]

    def plot(self):
      symplot(self.__Px[0] if self.__Px.ndim==1 else self.__Px[0,0])

```

```python
Px = Polynomial(np.matrix('0 2'), [0, -3])
Px.plot()
Px = Polynomial(np.matrix('-2 0 2'), [2, 0, 2])
Px.plot()
# Px.eval(0)
```
